### 2.0 Web Sracping Script<a id = "script"></a>

#### 2.1 Import package requests dan BeautifulSoup <a id = "import"></a>

```python
import requests
from bs4 import BeautifulSoup
```

#### 2.2 Request ke website <a id = "website"></a>

```python
page = requests.get("https://republika.co.id/")
```

### 2.3 Extract konten menjadi objek BeautifulSoup <a id = "extract"></a>

```python
obj = BeautifulSoup(page.text, 'html.parser');
```

## 3.0 Hasil Extract Konten <a id = "hasil"></a>

#### 3.1 Menampilkan Objek HTML <a id = "objek"></a>

```python
print ('Menampilkan objek html')
print ('=======================')
print (obj)
```

#### 3.2 Menampilkan Title Browser dengan Tag <a id = "title"></a>

```python
print ('\nMenampilkan title browser dengan tag')
print ('=======================================')
print (obj.title)
```

#### 3.3 Menampilkan Title Browser tanpa Tag <a id = "untitle"></a>

```python
print ('\nMenampilkan title browser tanpa tag')
print ('======================================')
print (obj.title.text)
```

#### 3.4 Menampilkan Semua Tag h2 <a id = "h2"></a>

```python
print ('\nMenampilkan semua tag h2')
print ('===========================')
print (obj.find_all('h2'))
```

#### 3.5 Menampilan Headline berdasarkan Div Class <a id = "head"></a>

```python
print ('\nMenampilkan headline berdasarkan div class')
print ('===========================================')
print (obj.find_all('div', class_='bungkus_txt_headline_center'))
```

#### 3.6 Menampilkan Semua Teks Headline <a id = "headline"></a>

```python
print ('\nMenampilkan semua teks headline')
print ('================================')
for headline in obj.find_all('div',class_='bungkus_txt_headline_center'):
    print (headline.find('h2').text)
```

### 4.0 Menyimpan dan Memproses ke Website <a id = "head4"></a>

#### 4.1 Menyimpan Data dalam File Text <a id = "file"></a>

```python
print ('\nMenyimpan headline pada file text')
print ('===================================')
f = open('headline.txt','a')
for headline in obj.find_all('div',class_='bungkus_txt_headline_center'):
    f.write(headline.find('h2').text)
    f.write('\n')
f.close
```
